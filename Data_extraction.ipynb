```python
# Library Importation

import os
import pandas as pd
from datetime import date, datetime
from dateutil.rrule import rrule, DAILY
import snscrape
```

```python
# Search term to look up in tweets
search_term = "health services"

# Maximum tweets per day
max_results = 1000
```

```python
# Google drive folder path

drive_folder_path = "/content/drive/MyDrive/twitter"
```

### Tweets in 2020

```python
# Start date
initial = date(2020, 1, 1)

# End date
final = date(2020,12 , 31)
```

```python
# Creating the list of dates

date_list = []
for dt in rrule(DAILY, dtstart=initial, until=final):
  temp_date = dt.strftime("%Y-%m-%d")
  date_list.append(temp_date)
```

```python
# Dataframe to append tweets

df_req_2020 = pd.DataFrame(columns=["content", "from_date", "end_date"])
```

```python
# Tweet extraction 

for i in range(len(date_list)-1):
  from_date = date_list[i]
  end_date = date_list[i+1]
  print("Executing date : " + date_list[i])
  extracted_tweets = "snscrape --format '{content!r}'" + f" --max-results {max_results} --since {from_date} twitter-search '{search_term} until:{end_date}' > extracted-tweets.txt"
  os.system(extracted_tweets)
  tweet_size = os.stat("extracted-tweets.txt").st_size
  if tweet_size > 0:
    temp_df = pd.DataFrame(columns=["content", "from_date", "end_date"])
    with open('extracted-tweets.txt') as f:
      contents = f.readlines()
      f.close()
    len_content = len(contents)
    from_date_list = [date_list[i]]*len_content
    end_date_list = [date_list[i+1]]*len_content
    temp_df["content"] = contents
    temp_df["from_date"] = from_date_list
    temp_df["end_date"] = end_date_list
    df_req_2020 = df_req_2020.append(temp_df, ignore_index=True)
    print("Successfully extracted : " + date_list[i])
```

```python
df_req_2020.to_excel(drive_folder_path + os.sep + "health_services_01_01_2020_31_12_2020.xlsx", index=False)
```
